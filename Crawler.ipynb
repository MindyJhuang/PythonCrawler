```python
import requests
from bs4 import BeautifulSoup

link = 'http://www.nownews.com/n/2014/09/03/1397623'
result = requests.get(link) 
response = result.text.encode('utf8')
soup = BeautifulSoup(response)
title = soup.find('h1', {"itemprop":"headline"}).text
firstLine = 1+len(soup.find('div', {"class":"story_content"}).find('p').text)
lastLine = -2-len(soup.find('div', {"class":"story_content"}).find('p',{"class":"bzkeyword"}).text)
story = soup.find('div', {"class":"story_content"}).text[firstLine:lastLine]

str=''
keywords = soup.find('div', {"class":"story_content"}).findAll('a',{'href':True})[1:]
for value in range (0, len(keywords)):
    str = str +keywords[value].text+ ' '

print link[25:35].replace('/','')
print title + story + str

print soup.find('div',{"class":"story_content"}).text
```

```python
import urlparse, requests, time
from bs4 import BeautifulSoup
link_file = open("link_list.txt", 'w') 
page_format = 'http://www.nownews.com/cat/pets/r/%d'
start = 1
for page in range(start,1000):
    result = requests.get(page_format%start) 
    response = result.text.encode('utf8')
    soup = BeautifulSoup(response)
    list_news = soup.find('ul', {"id":"result-list"}).findAll('div')
    if len(list_news)>0:
        list_news = soup.findAll('div', {"class":"box_1 clearfix"})
        for case in range(0,len(list_news)):            
            link = [tag['href'] for tag in list_news[case].findAll('a', {'href': True})][1]
            print link
            link_file.write(link + "\n") 
        start = start+1
        time.sleep(5)
    else:
        print "stop at page" + str(start)
        break
link_file.close() 
```
